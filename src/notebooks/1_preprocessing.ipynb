# Preprocesado y selección de los datos

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

```

- el dataset cuenta con una fecha, el texto preprocesado, su reduccion dimensional UMAP, el numero de palabras de cada sueño y 8 emociones basicas asociadas (tambien si es positivo o negativo)

```python

raw_df = pd.read_csv("../data/raw/dreams_syuzhet_df.csv",index_col=0).dropna()
raw_df = raw_df.reset_index(drop=True)
raw_df
```

# Sacamos los sueños por franjas educativas (high school, college, middle ages / adult 40s) categorias 0, 1, 2

```python
high_df = raw_df[raw_df['description'].str.contains('(?i)high school')].copy()
high_df["target"] = 0
high_df
```

```python
college_df = raw_df[raw_df['description'].str.contains('(?i)a college student|college women')].copy()
college_df["target"] = 1
college_df = college_df.iloc[:588,:]
college_df
```

```python
adult_df = raw_df[raw_df['description'].str.contains('(?i)middle-aged|Elizabeth')].copy()
adult_df["target"] = 2
adult_df = adult_df.iloc[:588,:]
adult_df
```

```python
all_df = pd.concat([high_df,college_df,adult_df],ignore_index=True)
all_df
```

# Separamos en train y test

```python
X = all_df.iloc[:,5:-1]
y = all_df.iloc[:,-1]
print(X.shape,y.shape)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)
train_df = pd.concat([X_train, y_train], axis=1)
test_df = pd.concat([X_test, y_test], axis=1)
print(train_df.shape,test_df.shape)
```

# Balanceamos las etiquetas de train

```python
# Balanceo etiquetas de train
tags = train_df['target'].value_counts()
print(tags)
```

```python
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(random_state=42)
X_rus, y_rus = rus.fit_resample(train_df.iloc[:,:-1], train_df['target'])
balanced_train = pd.concat([X_rus, y_rus], axis=1)
balanced_tags = balanced_train['target'].value_counts()
print(balanced_tags)
```

```python
# Guardados
#all_df.to_csv("../data/processed/dreams_ages.csv")
balanced_train.to_csv("../data/processed/balanced_train.csv")
test_df.to_csv("../data/processed/test.csv")
```
